````python
# Méthode d'Euler explicite pour la résolution de l'équation différentielle ordinaire(EDO)

# Méthode de Runge-Kutta 


import numpy as np
import matplotlib.pyplot as plt

u0=1                                      
t0=0                                      
t1=3            
N_24=24                                   
t_24=np.linspace(t0,t1,(N_24+1)) 

N_6=6
t_6=np.linspace(t0,t1,(N_6)+1)

print ("Le vecteur t est: ", t_24)
size=np.size(t_24)
print("La taille de t est: ", size)

# Solution Exacte

def solexact(t):
    y=np.exp(-4*t)
    return y

# Comparaison de la Méthode d'Euler explicite et la solution Exacte 

def F(t,u):
    return -4*u

# Euler Explicite(N=24)
    
def E(t,t1,u0):
    size=np.size(t)
    u=np.zeros(size)
    h=(t1)/(size-1)
    u[0]=u0

    for k in range(0,size-1):
        u[k+1]=u[k]+h*F(t[k],u[k])
    return u
            


# Le diagramme par la méthode de Runge-Kutta  

def Runge(t,t1,u0):
    
    N = len(t)
    h = float(t1/(N-1))
    u = np.zeros(N)
    u[0] = u0
    
    for i in range(0,N-1):
        k1 = h*F(t[i],u[i])
        k2 = h*F(t[i] + h/2, u[i] + k1/2)
        k3 = h*F(t[i] + h/2, u[i] + k2/2)
        k4 = h*F(t[i] + h, u[i] + k3)
        
        u[i+1] = u[i] + (1/6)*(k1 + 2*k2 + 2*k3 + k4)

    return u

U1 = Runge(t_24,t1,u0)
print("U1 :",U1)

U2 = Runge(t_6,t1,u0)
print("U2:",U2)
              
    
# Le diagramme d'Euler 

plt.plot(t_24,E(t_24,t1,u0),'o-')
plt.plot(t_24,solexact(t_24),'^-',)
plt.plot(t_6,E(t_6,t1,u0),'o-')
plt.grid()
plt.xlabel('t')
plt.ylabel('f(Ut)')
plt.legend(('Euler_t_24','Exact_t_24','Euler_t_6'),loc = 1)
plt.title("Méthode d'Euler explicite et la solution Exacte")
plt.show()

# Le diagramme de Runge-Kutta #####

plt.title("Methode Runge-Kutta")
plt.plot(t_24,Runge(t_24,t1,u0),'^-',linewidth = 0.5,label = 'N_24')
plt.plot(t_6,U2,'o-r',linewidth = 0.5, label = 'N_6')
plt.plot(t_24,solexact(t_24), 'o-',linewidth = 0.5, label = "Solution Exacte  N=24")
plt.xlabel("t")
plt.ylabel("u(t)")
plt.grid()
plt.legend()
plt.show()


````
